```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
print("step 1")
```

```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
```

```python
fig= plt.figure()
ax=plt.axes()

x=np.linspace(0, 5, 10)
print(x)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
ax.set_title('Simple plot')
ax.set_xlabel('wavelength')
ax.set_ylabel('amplitude')
plt.show()
```

```python
fig= plt.figure()
ax=plt.axes()

x=np.linspace(0, 10, 1000)
print(x)
ax.plot(x, np.sin(x), label='sin', color='pink')
ax.plot(x, np.cos(x), label='cos', color='red')
ax.plot(x, np.tan(x), label='tan')
ax.set_title('Simple plot')
ax.set_xlabel('wavelength')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()
```

```python
fig= plt.figure()
ax=plt.axes()

x=np.linspace(0, 10, 1000)
print(x)
ax.plot(x, np.sin(x), label='sin',linestyle='dashed')
ax.plot(x, np.cos(x), label='cos', linestyle='dotted')
ax.plot(x, np.sin(x+1), label='cos', linestyle='dashdot')
ax.set_title('Simple plot')
ax.set_xlabel('wavelength')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()
```
